```python

####  Question2 #####

def insert_key_values(n):
    list_pairs=[]
    for i in range(n):
        print('what is the student name:')
        a=input()
        print('what is her/his grade:')
        b=float(input())
        list_pairs.append((a,b))
    return dict(list_pairs)

### n is 3 here, you can define any number as "number of students":
my_dict=insert_key_values(3)    
print(my_dict)
print(my_dict.values())
```

```python
#### Question6_part I #####
import numpy as np
import time



##with normal distribution
a=np.random.randn(1,500000) 

## with uniform distribution from (0,1]
b=np.random.rand(1,500000)

## with uniform distribution from int values between [a,b]
c=np.random.randint(1,100, 500000)



#### Question6_Part II ####

begin1=time.time()
c_max1=np.max(c)
end1=time.time()

begin2=time.time()

c_max=-np.inf
for i in c:
    if c_max<i:
        c_max=i
    c_max2=c_max   
end2=time.time()

print(c_max1,c_max2)

print(f"Total runtime of the python built in method is {end1 - begin1} and for loop method is {end2-begin2}") 
```
